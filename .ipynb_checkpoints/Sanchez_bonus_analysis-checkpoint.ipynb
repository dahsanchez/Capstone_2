### Filtering, Sorting & Exploration

#### Create a list of all transactions from January 2024, sorted by sale amount from highest to lowest. Which transactions had a sale amount greater than $500? 

```python
StoreSales['Transaction Date'] = pd.to_datetime(StoreSales['Transaction Date'])
jan_data = StoreSales[(StoreSales['Transaction Date'] >= '2024-01-01') &(StoreSales['Transaction Date'] < '2024-02-01')
]

sorted_data = jan_data.sort_values(by='Sale Amount', ascending=False)
sorted_data
```

```python
above_500 = sorted_data[sorted_data['Sale Amount'] > 500]
above_500
```

#### Find all products whose product number begins with 10525. What category and subcategory do they belong to?

```python
ProdNumCat = Products.merge(ProductCat, on='CategoryID')
ProdNumCat = ProdNumCat[ProdNumCat['Prod Num'].str.startswith('10525')]
ProdCat = ProdNumCat[['Prod Num', 'Category', 'Subcategory']]
ProdCat
```

#### Which stores in your assigned territories had at least one transaction with a sale amount over $1,000? How many such transactions did each store have? 

```python
Florida = StoreDetail[StoreDetail['State'] == 'Florida']
FloridaStores= StoreSales[(StoreSales['Store ID'] >= 719) & (StoreSales['Store ID'] <=729)]

# Merge datasets
FLTransData = (
    FloridaStores
    .merge(Products, on='Prod Num')
    .merge(ProductCat, on='CategoryID')
    .groupby(['Store ID', 'Category'], as_index=False)
    .agg({
        'Sale Amount': 'sum',              # total sales
        'Prod Num': 'count'               # transaction count
    })
    .rename(columns={'Prod Num': 'Transaction Count'})
)
FLTransData
```
