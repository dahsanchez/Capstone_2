### Monthly total revenue for in-store sales in each of the two sales territories, over the full period covered by the data

#### Monthly Sales Data for Texas Territory

```python
StoreSales['Transaction Date'] = pd.to_datetime(StoreSales['Transaction Date']) # Makes transaction date a datetime data type
StoreSales['Year Month'] = StoreSales['Transaction Date'].dt.to_period('M') # month by month organization
TexasStores = StoreSales[(StoreSales['Store ID'] >= 901) & (StoreSales['Store ID'] <= 911)] # Only the Texas store IDs
TexasMonthly = (TexasStores.groupby('Year Month', as_index=False)['Sale Amount'].sum()) # Grouping by YearMonth and adding sale amount

TexasMonthly
```

#### Monthly Sales Data for the Florida Territory

```python
StoreSales['Transaction Date'] = pd.to_datetime(StoreSales['Transaction Date']) # Makes transaction date a datetime data type
StoreSales['Year Month'] = StoreSales['Transaction Date'].dt.to_period('M') # month by month organization
FloridaStores= StoreSales[(StoreSales['Store ID'] >= 719) & (StoreSales['Store ID'] <=729)] # only Florida stores
FloridaMonthly = (FloridaStores.groupby('Year Month', as_index=False)['Sale Amount'].sum()) # adding monthly sale amount, grouping the months togther

FloridaMonthly
```

### Ranking the sales performance of each store in each sales territory

#### Florida Stores ranked by top performing
#### The top proforming store are 
* 724(Miami)
* 728(Tallahassee)
* 722(Key West)

```python
FloridaStores= StoreSales[(StoreSales['Store ID'] >= 719) & (StoreSales['Store ID'] <=729)] # only using FLorida Stores
FloridaPerf = (FloridaStores.groupby('Store ID', as_index=False)['Sale Amount'].sum()) # total revenue by store
FloridaPerf = FloridaPerf.sort_values(['Sale Amount'], ascending=False) # top perfoming to least proforming
FloridaPerf = FloridaPerf.merge(Florida[['Store ID', 'Store Location']], on = 'Store ID') # Merging tables to get store location names
FloridaPerf.head()
```

```python
FLmax_value = max(FloridaPerf['Sale Amount']) # Highest perfoming store
FLmin_value = min(FloridaPerf['Sale Amount']) # lowest perorming store
plt.title("Store Preformance in FL") # title of chart
plt.bar(FloridaPerf['Store Location'], FloridaPerf['Sale Amount'], color=['green' if v == FLmax_value
                                                                          else 'red' if v == FLmin_value
                                                                          else 'mediumpurple'
                                                                          for v in FloridaPerf['Sale Amount'] ]) # to change color of bar to show
                                                                                                                # highest and lowest store
plt.xlabel('Store Locations')
plt.ylabel('Total Sale Amount')
plt.xticks(rotation=45, ha='right', fontsize=10) # make store names easy to read
plt.show()
plt.savefig('Store Preformace in FL.png')
```

This chart shows total sales revune per store in the Florida territorry with being the highest and Naples being th lowest.

#### Texas Stores ranked by top performing

### Analyze the number of transactions per month by product category within each assigned territory, along with the total monthly sales revenue by category, to identify the most popular products and uncover potential opportunities for growth.

#### Florida transaction count per month by category
#### The top 3 categories are:
* 2025-04	Art Supplies	330
* 2025-06	Stationery and Supplies	297
* 2025-05	Stationery and Supplies	297

```python
FloridaStores = FloridaStores.merge(StoreDetail, on='Store ID') # merging to get columns i need

FLTransData = (
    FloridaStores
    .merge(Products, on='Prod Num')
    .merge(ProductCat, on='CategoryID')
    .groupby(['State', 'Year Month', 'Category'])
    .size()
    .reset_index(name='Transaction Count')
)                                                   # Merging again to collect column names i need; showing monthly transaction count per category

FLTransData
```

```python
FLGrouped = FLTransData.groupby('Category')['Transaction Count'].sum().reset_index() # only using columns i neeed

FLCatMax = FLGrouped['Transaction Count'].max() # highest transaction count
FLCatMin = FLGrouped['Transaction Count'].min() # lowest transaction count
```

```python
plt.figure(figsize=(12, 6))
plt.title("Transactions by Category in FL") # title of bar chart

plt.bar(
    FLGrouped['Category'],
    FLGrouped['Transaction Count'],
    color=[
        'green' if v == FLCatMax
        else 'red' if v == FLCatMin
        else 'mediumpurple'
        for v in FLGrouped['Transaction Count'] # total number of transactions per category with the highest in green;lowest in red
    ]
)
plt.xlabel('Category')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45, ha='right', fontsize=10) # make the category names easy to read
plt.show()
plt.savefig('Transactions by Category in FL.png')
```

This chart shows over all transaction count per category for the Florida territorry with Stationery and Supplies being the highest and Books(General) being the lowest

```python
FLTransData.sort_values(by='Transaction Count', ascending=False).head(3) # showing top 3 hightest transaction counts
```

```python
FLTransData.sort_values(by='Transaction Count', ascending=False).tail(3) # showing bottom 3 transaction counts
```

#### Texas Transaction count per month 
#### The top 3 categories are:
* 2025-05	Stationery and Supplies	2937
* 2025-06	Stationery and Supplies	2816
* 2025-10	Stationery and Supplies	2618


```python
TexasStores = TexasStores.merge(StoreDetail, on='Store ID') # merging to get columns i need

TXTransData = (
    TexasStores
    .merge(Products, on='Prod Num')
    .merge(ProductCat, on='CategoryID')
    .groupby(['State', 'Year Month', 'Category'])
    .size()
    .reset_index(name='Transaction Count') # meringing to get columns i need; showing monthly transaction count per category
)

TXTransData
```

```python
TXGrouped = TXTransData.groupby('Category')['Transaction Count'].sum().reset_index() # only using columns i need

TXCatMax = TXGrouped['Transaction Count'].max() # highest transaction count
TXCatMin = TXGrouped['Transaction Count'].min() # lowest transaction count
```

```python
plt.figure(figsize=(12, 6))
plt.title("Transactions by Category in TX") # title of bar chart
plt.bar(
    TXGrouped['Category'],
    TXGrouped['Transaction Count'],
   color=[
        'green' if v == TXCatMax
        else 'red' if v == TXCatMin
        
        else 'mediumpurple'
        for v in TXGrouped['Transaction Count']]  # shows total transaction count per category; highest in green; lowest in red
    
)
plt.xlabel('Category')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45, ha='right', fontsize=10) # makes the category names easier to read
plt.show()
plt.savefig('Transactions by Category in TX.png')
```

This chart shows over all transaction count per category for the Texas with Stationery and Supplies being the highest and Books(General) being the lowest
