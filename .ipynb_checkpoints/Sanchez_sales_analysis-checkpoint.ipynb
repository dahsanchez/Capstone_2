#  Core Marketing Analysis

### Identify territory managers, store IDs, and store locations for each assigned territory.

```python
Texas = StoreDetail[(StoreDetail['State'] == 'Texas')] # only showing Texas stores
Texas

# The manager for Florida is Lana Ilana with stores 719-729
# The manager for Texas is Jeff "Howdy" Richards with store 901-911
```

```python
Florida = StoreDetail[(StoreDetail['State'] == 'Florida')] # only showing Florida stores
Florida
```

### Monthly total revenue for in-store sales in each of the two sales territories, over the full period covered by the data

#### Monthly Sales Data for Texas Territory

```python
StoreSales['Transaction Date'] = pd.to_datetime(StoreSales['Transaction Date']) # Makes transaction date a datetime data type
StoreSales['Year Month'] = StoreSales['Transaction Date'].dt.to_period('M') # month by month organization
TexasStores = StoreSales[(StoreSales['Store ID'] >= 901) & (StoreSales['Store ID'] <= 911)] # Only the Texas store IDs
TexasMonthly = (TexasStores.groupby('Year Month', as_index=False)['Sale Amount'].sum()) # Grouping by YearMonth and adding sale amount

TexasMonthly
```

#### Monthly Sales Data for the Florida Territory

```python
StoreSales['Transaction Date'] = pd.to_datetime(StoreSales['Transaction Date']) # Makes transaction date a datetime data type
StoreSales['Year Month'] = StoreSales['Transaction Date'].dt.to_period('M') # month by month organization
FloridaStores= StoreSales[(StoreSales['Store ID'] >= 719) & (StoreSales['Store ID'] <=729)] # only Florida stores
FloridaMonthly = (FloridaStores.groupby('Year Month', as_index=False)['Sale Amount'].sum()) # adding monthly sale amount, grouping the months togther

FloridaMonthly
```

### Ranking the sales performance of each store in each sales territory

#### Florida Stores ranked by top proforming
#### The top proforming store are 
* 724(Miami)
* 728(Tallahassee)
* 722(Key West)

```python
FloridaStores= StoreSales[(StoreSales['Store ID'] >= 719) & (StoreSales['Store ID'] <=729)] # oblyinh using FLorida Stores
FloridaPerf = (FloridaStores.groupby('Store ID', as_index=False)['Sale Amount'].sum()) # total reveue by store
FloridaPerf = FloridaPerf.sort_values(['Sale Amount'], ascending=False) # top preofoming to least proforming
FloridaPerf = FloridaPerf.merge(Florida[['Store ID', 'Store Location']], on = 'Store ID')
FloridaPerf
```

#### Texas Stores ranked by top proforming
#### The top proforming store are 
* 905(Beaumont)
* 910(Fort Worth)
* 909(Desoto)

```python
TexasStores = StoreSales[(StoreSales['Store ID'] >= 901) & (StoreSales['Store ID'] <= 911)]
TexasPerf = (TexasStores.groupby('Store ID', as_index=False)['Sale Amount'].sum()) # total reveue by store
TexasPerf = TexasPerf.sort_values(['Sale Amount'], ascending=False) # top preofoming to least proforming
TexasPerf =  TexasPerf.merge(Texas[['Store ID', 'Store Location']], on = 'Store ID') # show location name with store id 
TexasPerf
```
